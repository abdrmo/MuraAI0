```python

def save_to_gdrive(file_name='feature_names_pid02072025.csv',
                  dest_folder='/content/drive/MyDrive/'):


    # Set source path (always in /content/)
    source_path = f'/content/{file_name}'

    # Ensure destination folder ends with slash
    dest_folder = dest_folder if dest_folder.endswith('/') else dest_folder + '/'
    destination_path = f'{dest_folder}{file_name}'

    # Ensure destination folder ends with slash
    dest_folder = dest_folder if dest_folder.endswith('/') else dest_folder + '/'
    destination_path = f'{dest_folder}{file_name}'

    try:
        # Verify source exists
        if not os.path.exists(source_path):
            raise FileNotFoundError(f"Source file not found at {source_path}")

        # Copy the file
        shutil.copy(source_path, destination_path)
        print(f"File successfully copied from {source_path}")
        print(f"Saved to Google Drive at: {destination_path}")
        return True
    except Exception as e:
        print(f"Error: {str(e)}")
        return False

save_to_gdrive()
```

```python
from google.colab import drive
drive.mount('/content/drive')

import pickle
import sys
import inspect

def get_saveable_variables():
    """Get all variables that can be pickled"""
    saveable = {}
    for name, value in globals().items():
        # Skip special variables, modules, and functions
        if name.startswith('_'):
            continue
        if inspect.ismodule(value):
            continue
        if callable(value):
            continue

        # Try to pickle to test if it's saveable
        try:
            pickle.dumps(value)
            saveable[name] = value
        except (pickle.PicklingError, TypeError):
            print(f"Skipping {name} (type: {type(value)}) - not picklable")
    return saveable

# Get saveable variables
saveable_vars = get_saveable_variables()

# Save to Drive
with open('/content/drive/My Drive/all_variables.pkl', 'wb') as f:
    pickle.dump(saveable_vars, f)

print(f"Successfully saved {len(saveable_vars)} variables:")
print("\n".join(saveable_vars.keys()))
```
