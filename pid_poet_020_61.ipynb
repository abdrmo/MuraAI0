```python
# @title S0-2-0__W21__C0__defFilterKin

def filter_kinematic_and_ml_vars(df, include_keywords=None, exclude_keywords=None, min_non_nan_ratio=0.8):
    """
    Filters a DataFrame for ML-relevant and kinematic variables.

    Parameters:
        df (pd.DataFrame): The original DataFrame from ROOT.
        include_keywords (list): Additional substrings to include.
        exclude_keywords (list): Substrings to exclude (e.g., ['present', 'emptyEvent']).
        min_non_nan_ratio (float): Minimum fraction of non-NaN values to keep a column.

    Returns:
        pd.DataFrame: Cleaned and filtered DataFrame.
    """
    # Base keywords
    default_keywords = ['pt', 'eta', 'phi', 'rho', 'softElectronByPt', 'softMuonByPt']
    if include_keywords:
        default_keywords += include_keywords
    if exclude_keywords is None:
        exclude_keywords = ['present', 'emptyEvent', 'ref_', 'product_']

    # Step 1: Column selection
    selected_cols = [
        col for col in df.columns
        if any(k.lower() in col.lower() for k in default_keywords)
        and not any(ex.lower() in col.lower() for ex in exclude_keywords)
    ]

    # Step 2: Create filtered DF
    filtered_df = df[selected_cols].copy()

    # Step 3: Remove columns with too many NaNs/infs
    clean_cols = [
        col for col in filtered_df.columns
        if (filtered_df[col].replace([np.inf, -np.inf], np.nan).notna().mean() >= min_non_nan_ratio)
    ]
    filtered_df = filtered_df[clean_cols]

    # Step 4: Drop rows with NaNs or infs
    filtered_df.replace([np.inf, -np.inf], np.nan, inplace=True)
    filtered_df.dropna(inplace=True)

    return filtered_df
    # ACESS fastject reco, u52
    # ----------
```

```python
import ast
# import numpy as np
# import pandas as pd
from collections import defaultdict

# Dictionary to track failure counts
failure_counts = defaultdict(int)

def safe_string_to_array(val):
    if isinstance(val, str):
        val = val.strip()

        # hmm.. literal_eval first, .. a mistake i guess
        try:
            parsed = ast.literal_eval(val)
            if isinstance(parsed, (list, tuple)):
                return np.array(parsed, dtype=np.float32)
        except (ValueError, SyntaxError):
            failure_counts['literal_eval'] += 1

        # Try np.fromstring
        try:
            arr = np.fromstring(val.strip("[]"), sep=' ')
            if arr.size > 0:
                return arr.astype(np.float32)
            else:
                failure_counts['fromstring_empty'] += 1
        except Exception:
            failure_counts['fromstring_exception'] += 1

    elif isinstance(val, (list, tuple, np.ndarray)):
        try:
            return np.array(val, dtype=np.float32)
        except Exception:
            failure_counts['array_coercion'] += 1

    # Log as generic failure
    failure_counts['unhandled'] += 1
    return np.nan
```

```python

def save_to_gdrive(file_name='feature_names_pid02072025.csv',
                  dest_folder='/content/drive/MyDrive/'):


    # Set source path (always in /content/)
    source_path = f'/content/{file_name}'

    # Ensure destination folder ends with slash
    dest_folder = dest_folder if dest_folder.endswith('/') else dest_folder + '/'
    destination_path = f'{dest_folder}{file_name}'

    # Ensure destination folder ends with slash
    dest_folder = dest_folder if dest_folder.endswith('/') else dest_folder + '/'
    destination_path = f'{dest_folder}{file_name}'

    try:
        # Verify source exists
        if not os.path.exists(source_path):
            raise FileNotFoundError(f"Source file not found at {source_path}")

        # Copy the file
        shutil.copy(source_path, destination_path)
        print(f"File successfully copied from {source_path}")
        print(f"Saved to Google Drive at: {destination_path}")
        return True
    except Exception as e:
        print(f"Error: {str(e)}")
        return False

save_to_gdrive()
```
